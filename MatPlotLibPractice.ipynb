```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(0,20)
y = np.arange(5,25)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.title('Practice Scatter Plot')
plt.savefig('PracticeScatter.jpg')
```

```python
y=1/(x+1)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r:')
plt.subplot(2,2,2)
plt.plot(x,1/y,'ro--')
plt.subplot(2,2,3)
plt.plot(x,y*y,'r*--')
```

```python
x = np.arange(0,10*np.pi,0.001)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = [2,4,6,8,10]
y = [1,3,5,7,9]
x2 = [2.5,4.5,6.5,8.5,10.5]
y2 = [11,13,15,17,19]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

```python
a = np.array([2,4,6,8,10,1,3,5,7,9,2.5,4.5,6.5,8.5,10.5,11,13,15,17,19])
plt.hist(a,bins = 20)
plt.show()
```

```python
data = [np.random.normal(0,std,100) for std in range(1,9)]

plt.boxplot(data,vert=True , patch_artist='True')
```
